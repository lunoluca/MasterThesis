```python
import numpy as np
import matplotlib.pyplot as plt

# Years and baseline energy consumption (EJ/year)
years = np.array([2020, 2025, 2030, 2035, 2040])
baseline_energy = {
    'BAU': np.array([13.9591, 13.9315, 14.0723, 14.4057, 14.8193]),
    'EDCE': np.array([13.9855, 13.809, 13.5972, 13.6341, 13.4658]),
    'TTT': np.array([13.9807, 13.6972, 13.3978, 13.0653, 12.6207])
}

# Automation & Robotics impact (adoption rate × 25%) per scenario
ai_impact = {
    'BAU': np.array([0, 0.0557, 0.0744, 0.0975, 0.125]),
    'EDCE': np.array([0, 0.0586, 0.0929, 0.1261, 0.15]),
    'TTT': np.array([0, 0.061, 0.1152, 0.1726, 0.2125])
}

# Plot setup
fig, axs = plt.subplots(1, 3, figsize=(18, 6), sharey=True)
scenarios = ['BAU', 'EDCE', 'TTT']

for i, scenario in enumerate(scenarios):
    baseline = baseline_energy[scenario]
    impact = ai_impact[scenario]
    reduced = baseline * (1 - impact)

    axs[i].plot(years, baseline, label='Baseline', linestyle='--', color='red', linewidth=2)
    axs[i].plot(years, reduced, label='With Automation & Robotics Impact', color='blue', linewidth=2)
    axs[i].fill_between(years, reduced, baseline, color='blue', alpha=0.3)
    axs[i].set_title(f'{scenario} Scenario')
    axs[i].set_xlabel('Year')
    axs[i].grid(True)
    axs[i].set_xticks(years)

axs[0].set_ylabel('Energy Consumption (EJ/year)')
fig.suptitle('Automation & Robotics Impact on Energy Consumption (2020–2040)', fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.97])
plt.ylim(0, 17) 
plt.show()
```
