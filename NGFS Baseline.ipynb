```python
import matplotlib.pyplot as plt

# Define year labels
years = [2020, 2025, 2030, 2035, 2040]

# CO emissions data (Mt CO2/year)
co2_emissions = {
    'BAU (Current Policies)': [209.6612, 176.6175, 151.7375, 135.3388, 120.4279],
    'EDCE (Below 2C)':       [209.6612, 173.936,   142.4242, 115.7948, 75.2439],
    'TTT (Net Zero 2050)':    [209.6612, 171.657,   139.8397, 111.4298, 66.9889],
}

# Final energy consumption data (EJ/year)
energy_consumption = {
    'BAU (Current Policies)': [13.9591, 13.9315, 14.0723, 14.4057, 14.8193],
    'EDCE (Below 2C)':       [13.9855, 13.809, 13.5972, 13.6341, 13.4658],
    'TTT (Net Zero 2050)':    [13.9807, 13.6972, 13.3978, 13.0653, 12.6207],
}

# -------------------- CO Emissions Chart --------------------
plt.figure(figsize=(10, 6))
for label, values in co2_emissions.items():
    plt.plot(years, values, marker='o', label=label)
plt.title("Baseline CO Emissions for EU27 (20202040)")
plt.xlabel("Year")
plt.ylabel("CO Emissions (Mt/year)")
plt.xticks(ticks=years)  # Force integer years
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.ylim(0, 250) 
plt.show()
plt.savefig('CO2 Emissions Baseline')

# -------------------- Energy Consumption Chart --------------------
plt.figure(figsize=(10, 6))
for label, values in energy_consumption.items():
    plt.plot(years, values, marker='o', label=label)
plt.title("Baseline Final Energy Consumption for EU27 Industry (20202040)")
plt.xlabel("Year")
plt.ylabel("Energy Consumption (EJ/year)")
plt.xticks(ticks=years)  # Force integer years
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.ylim(0, 17) 
plt.show()
plt.savefig('Energy Consumption Emissions Baseline')
```
